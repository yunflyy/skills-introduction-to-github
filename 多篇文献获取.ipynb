```python
import requests
from lxml import etree
import pandas as pd
```

```python
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
papers = []
colIndex = ['title', 'authors', 'pmID', 'mag', 'year', 'abstract', 'kw']
def getSingleInfo(url):
    singleList = []
    rsp = requests.get(url, headers=headers, timeout=500).text
    html = etree.HTML(rsp)
    title = html.xpath('//*[@id="full-view-heading"]/h1/text()')
    title = title[0].strip()
    print(title)
    singleList.append(title)
    authors = html.xpath('//*[@id="full-view-heading"]/div[2]/div/div/span/a/text()')
    authors = ','.join(authors) 
    print(authors)
    singleList.append(authors)
    pmID = html.xpath('//*[@id="full-view-identifiers"]/li[1]/span/strong/text()')
    pmID = pmID[0]
    print(pmID)
    singleList.append(pmID)
    mag = html.xpath('//*[@id="full-view-journal-trigger"]/text()')
    mag = mag[0].strip()
    print(mag)
    singleList.append(mag)
    info = html.xpath('//*[@id="full-view-heading"]/div[1]/div/span[2]/text()')
    year = info[0][0:4]
    print(year)
    singleList.append(year)
    abstract = html.xpath('//*[@id="eng-abstract"]/p/text()')
    abstract = abstract[0].strip()
    print(abstract)
    singleList.append(abstract)
    try:
        kw = html.xpath('//*[@id="abstract"]/p/text()')[1].strip()
        print(kw)
        singleList.append(kw)
    except:
        pass
    #print(singleList)
    papers.append(singleList)
    
```

```python
def getLink(url):
    rsp = requests.get(url, headers=headers, timeout=500).text
    html = etree.HTML(rsp)
    for i in range(1, 11):
        link = html.xpath(f'//*[@id="search-results"]/section[1]/div[2]/div/article[{i}]/div[2]/div[1]/a/@href')
        link = 'https://pubmed.ncbi.nlm.nih.gov' + link[0]
        print(link)
        getSingleInfo(link)
        
        print('\n\n=================================\n\n')

for p in range(1, 2):
    url = f'https://pubmed.ncbi.nlm.nih.gov/?term=atherosclerosis&filter=datesearch.y_5&filter=simsearch1.fha&filter=simsearch3.fft&page={p}'
    print(f'这是第{p}页')
    getLink(url)
excel = pd.DataFrame(papers, columns=colIndex)
excel.to_excel('PudMed文献信息.xlsx')
         
```
